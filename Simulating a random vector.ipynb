```python
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
u_1 = [random.uniform(0,1) for i in range(5000)]
u_2 = [random.uniform(0,1) for i in range(5000)]

x_1, x_2 = [], []

for i in range(5000):
    if (u_1[i] <= 0.5 and u_2[i] <= 2 * u_1[i]) or \
    (u_1[i] > 0.5 and u_2[i] <= -2 * u_1[i] + 2):
        x_1.append(u_1[i])
        x_2.append(u_2[i])
```

```python
plt.plot(x_1[:1000], x_2[:1000], 'o', color='black');
```
